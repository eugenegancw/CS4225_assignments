Pre-Processing data for weather and Trip

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
from pyspark.sql.functions import col, to_date
```

```python
# Specify the DBFS FileStore folder
filestore_folder1 = 'dbfs:/FileStore/training_data/TripData'
filestore_folder2= 'dbfs:/FileStore/testing_data/TripData'

# List all files in the FileStore folder
file_paths1 = dbutils.fs.ls(filestore_folder1)
file_paths2 = dbutils.fs.ls(filestore_folder2)
file_paths1.extend(file_paths2) #merge them into one list

aggregate_df = None
# Iterate through each file and read it into a DataFrame
for file_path in file_paths1:
    if file_path.isFile() and file_path.name.endswith('.csv'):
        # print(file_path.path)
        file_df = spark.read.option("header", "true").csv(file_path.path)
        file_df = file_df.withColumn("started_at", to_date(col("started_at").cast("timestamp")))
        daily_aggregated_df = file_df.groupBy("started_at").count()
        # Union DataFrames
        if aggregate_df is None:
            aggregate_df = daily_aggregated_df
        else:
            aggregate_df = aggregate_df.union(daily_aggregated_df)
```

```python
aggregate_df2 = aggregate_df.withColumnRenamed("started_at", "datetime").withColumnRenamed("count", "trip_count")
```

```python
#downloaded the weather data and has already uploaded it into the weatehr_filePath
weather_filePath='/FileStore/weather_data.csv'
weather_df = spark.read.csv(weather_filePath, header="true", escape= "\t") 
```

```python
noNullsDF = weather_df.na.drop(subset=["temp"])
display(noNullsDF)
```

```python
display(weather_df.describe())
weather_df.dtypes
```

```python
# Testing
from pyspark.sql.functions import col, unix_timestamp, date_format
from pyspark.sql.types import DoubleType

# Convert Strings to Timestamp
weather_df = weather_df.withColumn('datetime', weather_df['datetime'].cast('timestamp'))
weather_df = weather_df.withColumn('sunrise', unix_timestamp(weather_df['sunrise'], 'yyyy-MM-dd\'T\'HH:mm:ss').cast('timestamp'))
weather_df = weather_df.withColumn('sunset', unix_timestamp(weather_df['sunset'], 'yyyy-MM-dd\'T\'HH:mm:ss').cast('timestamp'))

# Extract Day of the Week
weather_df = weather_df.withColumn('day_of_week', date_format(col('datetime'), 'E'))

# Calculate Daylight Duration
weather_df = weather_df.withColumn('daylight_duration', (col('sunset').cast('double') - col('sunrise').cast('double')))
weather_df = weather_df.withColumn('daylight_hours', (col('daylight_duration') / 3600))
```

```python
#select the variabels we want to keep for weather data
columnsToKeep = [
  "datetime",
  "temp",
  "feelslike",
  "humidity",
  "precip",
  "snow",
  "windspeed",
  "visibility",
  "solarenergy",
  "uvindex",
  "severerisk",
  "description", "day_of_week",  "daylight_hours"]

fil_weather_df = weather_df.select(columnsToKeep)
fil_weather_df.cache().count()
```

```python
# Convert multiple columns from string to double
numerical_cols = ["temp",
  "feelslike",
  "humidity",
  "precip",
  "snow",
  "windspeed",
  "visibility",
  "solarenergy",
  "uvindex",
  "severerisk", "daylight_hours"] # can consider variables such as sunset and sunrise if not good enough

for column in numerical_cols:
    fil_weather_df = fil_weather_df.withColumn(column, col(column).cast("double"))
```

```python
from pyspark.sql.functions import sum

# Identify string and double columns
string_cols = [c for c in fil_weather_df.columns if 'string' in fil_weather_df.schema[c].dataType.typeName()] # check if day_of_week is inside
# Count null values for each string column
null_counts_string = fil_weather_df.select([sum(col(c).isNull().cast("int")).alias(c) for c in string_cols])

# Count null values for each double column
null_counts_double = fil_weather_df.select([sum(col(c).isNull().cast("int")).alias(c) for c in numerical_cols])

# Show the null counts summary for string columns
print("Null counts for each string column:")
display(null_counts_string)

# Show the null counts summary for double columns
print("Null counts for each double column:")
display(null_counts_double)
```

```python
from pyspark.sql.functions import when
#only choose those columns that have at least 1 null value as imputeCols
imputeCols = ["severerisk"]

for c in imputeCols:
  fil_weather_df = fil_weather_df.withColumn(c + "_na", when(col(c).isNull(), 1.0).otherwise(0.0))
```

#### Read file from saved DataPath

```python
# Split the table for the training and testing data
dataPath = "dbfs:/FileStore/cleaned_bike_final.csv"
bike_df = spark.read.csv(dataPath, header=True, inferSchema=True)
```

#### Filtering and splitting data into training and testing data

```python
from pyspark.sql.functions import col, to_date, lit

# Define the filter condition
filter_ = col("datetime") >= lit("2023-01-01")

# Apply the filter to create two DataFrames
testing_bike_df = bike_df.filter(filter_)
training_bike_df = bike_df.filter(~filter_)
```

## 2. Build ML Pipeline using Training Data

```python
from pyspark.ml.feature import Imputer, RFormula
from pyspark.ml import Pipeline
from pyspark.ml.regression import LinearRegression

# Assuming imputeCols is a list of numerical columns to be imputed
imputer = Imputer(strategy="median", inputCols=imputeCols, outputCols=imputeCols)
# selected_variables=['temp','feelslike','humidity','precip','snow','windspeed','visibility','solarenergy','uvindex','severerisk','description', #'day_of_week','daylight_hours']
exclude_variables = ["severerisk_na", "datetime"] #remove the na_columns and datetime as features to predict the trip_count
#check if there is any additioanl columns
exclude_string = " ".join([f"- {var}" for var in exclude_variables])
print(exclude_string)
# Define RFormula; RFormula will also treats columns of type String as a categorical feature and string indexes & one hot encodes it for us. 
rFormula = RFormula(formula=f"trip_count ~ . {exclude_string}" , featuresCol="features", labelCol="trip_count", handleInvalid="skip")

# Define the Linear Regression model
lr = LinearRegression(labelCol="trip_count", featuresCol="features")
```

```python
# Create the Pipeline
pipeline = Pipeline(stages=[imputer, rFormula, lr])

# Fit the Pipeline on the training data
pipelineModel = pipeline.fit(training_bike_df)

#additional steps
pred_train= pipelineModel.transform(training_bike_df)

pred_train.select("features", "trip_count", "prediction").show(5)
```

## 3. Evaluate ML Pipeline using Testing Data

```python
# Transform the test data
pred_test = pipelineModel.transform(testing_bike_df)
pred_test.select("features", "trip_count", "prediction").show(5)
```

```python
#Evaluation
from pyspark.ml.evaluation import RegressionEvaluator

regressionEvaluator = RegressionEvaluator(predictionCol="prediction", labelCol="trip_count")
rmse = round(regressionEvaluator.setMetricName("rmse").evaluate(pred_test), 2)

r2 = round(regressionEvaluator.setMetricName("r2").evaluate(pred_test), 2)

mae = round(regressionEvaluator.setMetricName("mae").evaluate(pred_test), 2)
print(f"RMSE is {rmse}")
print(f"R2 is {r2}")
print(f"MAE is {mae}")
```
